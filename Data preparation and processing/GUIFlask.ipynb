```python
from flask import Flask, request, render_template_string, redirect, url_for
import os

app = Flask(__name__)

# Path to the directory where uploaded PDFs will be stored
UPLOAD_FOLDER = 'pdf_upload_gui'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

# Ensure the upload directory exists
if not os.path.exists(UPLOAD_FOLDER):
    os.makedirs(UPLOAD_FOLDER)

# HTML template for the form with added styling and additional fields
form_html = '''
<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <title>Job Searcher</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      background-color: #e8f4f8;
      display: flex;
      flex-direction: column;
      align-items: center;
      height: 100vh;
      margin: 0;
      padding: 20px;
    }
    h1 {
      color: #333333;
      margin-bottom: 20px;
    }
    .container {
      text-align: center;
      background-color: #ffffff;
      padding: 20px;
      border-radius: 10px;
      box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
      width: 100%;
      max-width: 500px;
      margin-top: 20px;
    }
    form {
      margin-top: 20px;
    }
    label, input[type="text"], input[type="email"], input[type="file"], input[type="submit"] {
      display: block;
    }
    input[type="text"], input[type="email"], input[type="file"], input[type="submit"] {
      padding: 10px;
      width: 80%;
      box-sizing: border-box;
      margin: 10px auto;
    }
    input[type="submit"] {
      background-color: #4CAF50;
      color: white;
      border: none;
      border-radius: 5px;
      cursor: pointer;
      margin-top: 20px; /* Added spacing above the button */
      margin-bottom: 20px; /* Added spacing below the button */
    }
    input[type="submit"]:hover {
      background-color: #45a049;
    }
    label[for="pdf"] {
      margin-bottom: 10px; /* Added margin below the label for the PDF file */
    }
  </style>
</head>
<body>
<h1>Welcome to your job searching application</h1>
<div class="container">
  <h2>Fill in the following fields to match with a job</h2>
  <form action="/upload" method="post" enctype="multipart/form-data">
    <label for="name">Name:</label>
    <input type="text" id="name" name="name" required>
    <label for="email">Email:</label>
    <input type="email" id="email" name="email" required>
    <label for="pdf">PDF file:</label>
    <div>
      <input type="file" id="pdf" name="pdf" accept="application/pdf" style="display:none" onchange="document.getElementById('fileLabel').innerText = this.files[0].name;">
      <button type="button" onclick="document.getElementById('pdf').click()">Choose file</button>
      <span id="fileLabel"></span>
    </div>
    <input type="submit" value="Get your job">
  </form>
</div>
</body>
</html>
'''

# Route to display the form
@app.route("/")
def form():
    return render_template_string(form_html)

# Route to handle file upload
@app.route("/upload", methods=["POST"])
def upload():
    if 'pdf' not in request.files:
        return "<p>No file part</p>"
    file = request.files['pdf']
    if file.filename == '':
        return "<p>No selected file</p>"
    if file and file.filename.endswith('.pdf'):
        name = request.form['name']
        email = request.form['email']
        file.save(os.path.join(app.config['UPLOAD_FOLDER'], file.filename))
        return f"<p>Thank you, {name}. Your CV has been uploaded.</p><p>We will contact you at {email}.</p>"
    else:
        return "<p>Invalid file type. Please upload a PDF file.</p>"

# Step 3: Run the Flask app in a separate thread
import threading

def run_flask():
    app.run(debug=True, use_reloader=False)

flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```
