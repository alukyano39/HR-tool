```python
# Load VDAB data
vdab_file_path = 'C:/Users/Cool_/Documents/GitHub/PPA/hr_tool_irakli/vdab_categories_provinces.csv'
vdab_jobs = pd.read_csv(vdab_file_path)
```

```python
import re

# Function to classify the job as starter or experienced
def classify_job(description, title):
    description = description.lower()
    title = title.lower()
    
    starter_keywords = ['no experience required', 'junior', 'entry-level', 'starter', 'graduate', 'young']
    experienced_keywords = ['years of experience', 'senior', 'experienced', 'minimal', 'at least', 'no less than']
    
    for keyword in starter_keywords:
        if re.search(keyword, description) or re.search(keyword, title):
            return 'starter'
    
    for keyword in experienced_keywords:
        if re.search(keyword, description) or re.search(keyword, title):
            return 'experienced'
    
    return 'experienced'  # Default to experienced if no keywords are found

# Apply the function to the translated description
vdab_jobs['job_level'] = vdab_jobs.apply(lambda x: classify_job(x['description_translated'], x['title_translated']), axis=1)

# Display the updated dataframe
vdab_jobs
```
