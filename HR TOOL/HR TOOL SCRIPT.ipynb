```python
import os
import re
import unicodedata
import pandas as pd
from flask import Flask, render_template, request, redirect, url_for, session, flash
from werkzeug.utils import secure_filename
from bs4 import BeautifulSoup
from itsdangerous import URLSafeTimedSerializer, SignatureExpired
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
import smtplib
import fitz  # PyMuPDF
from sentence_transformers import SentenceTransformer, util
import time
import spacy
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
import torch
import nltk
from nltk.corpus import stopwords
from sklearn.preprocessing import normalize
from scipy.special import rel_entr

app = Flask(__name__, static_url_path='/static', static_folder='static')
app.secret_key = os.urandom(24)  # Secret key for session management

# Email confirmation setup
s = URLSafeTimedSerializer(app.secret_key)

# Temporary in-memory storage for users
users = {
    'admin@admin.com': {
        'first_name': 'Admin',
        'last_name': 'User',
        'address': 'Admin Address',
        'postal_code': '12345',
        'password': 'password',
        'confirmed': True
    }
}
confirmations = {}

# Load VDAB data
vdab_file_path = 'VDAB_categories_provinces_level.csv'
vdab_jobs = pd.read_csv(vdab_file_path)
vdab_jobs['text'] = vdab_jobs['title_translated'].fillna('') + ' ' + vdab_jobs['employer'].fillna('') + ' ' + vdab_jobs['location'].fillna('') + ' ' + vdab_jobs['description_translated'].fillna('')

# Load SBERT model with GPU support if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer('sentence-transformers/all-MiniLM-L12-v2', device=device)

# Function to clean text
def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    # Normalize Unicode characters
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Remove @mentions, URLs, hashtags, non-ASCII characters, and special characters
    text = re.sub(r"@\w+|https?://\S+|#\w+|[^\x00-\x7F]+|[^\w\s]", "", text)
    
    # Remove newline characters, leading/trailing spaces, multiple spaces, and numbers
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"^\s+|\s+$", "", text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"\b\d+\b", "", text)
    
    # Convert to lowercase
    text = text.lower()
    
    return text

# Clean the job descriptions
vdab_jobs['clean_text'] = vdab_jobs['text'].apply(clean_text)

def extract_text_from_pdf(pdf_file_path):
    doc = fitz.open(pdf_file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    text = re.sub(r'\s+', ' ', text)  # Replace multiple whitespace with a single space
    text = text.strip()
    return clean_text(text)

def calculate_kl_divergence(jobs, query_text):
    # Encode jobs in larger batches
    batch_size = 128
    job_embeddings = model.encode(jobs, convert_to_tensor=True, batch_size=batch_size, show_progress_bar=True)
    query_embedding = model.encode([query_text], convert_to_tensor=True, batch_size=1, show_progress_bar=True)

    # Ensure embeddings are valid probability distributions
    job_embeddings = torch.nn.functional.softmax(job_embeddings, dim=1)
    query_embedding = torch.nn.functional.softmax(query_embedding, dim=1)

    # Calculate KL divergence
    kl_divergences = torch.nn.functional.kl_div(query_embedding.log(), job_embeddings, reduction='none').sum(dim=1)
    return kl_divergences

def find_similar_jobs(jobs, query_text, keyword=None, top_n=10, preferred_provinces=None, preferred_categories=None, experience_levels=None):
    if preferred_provinces:
        preferred_provinces = [province.lower() for province in preferred_provinces]

    # Filter VDAB jobs based on keyword, province, categories, and experience level
    filtered_jobs = jobs.copy()
    if keyword:
        filtered_jobs = filtered_jobs[(filtered_jobs['title_translated'].str.contains(keyword, case=False, na=False)) | 
                                      (filtered_jobs['description_translated'].str.contains(keyword, case=False, na=False))]
    if preferred_provinces and 'province' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['province'].str.lower().str.contains('|'.join(preferred_provinces), case=False, na=False)]
    if preferred_categories and 'categories' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['categories'].apply(lambda x: any(cat in preferred_categories for cat in x.split(',')))]
    if experience_levels:
        filtered_jobs = filtered_jobs[filtered_jobs['job_level'].str.contains('|'.join(experience_levels), case=False, na=False)]

    filtered_jobs['clean_text'] = filtered_jobs['clean_text'].apply(clean_text)  # Ensure clean_text is applied

    # Calculate KL divergence for all job embeddings
    kl_divergences = calculate_kl_divergence(filtered_jobs['clean_text'].tolist(), query_text)
    filtered_jobs['kl_divergence'] = kl_divergences.cpu().numpy()
    filtered_jobs = filtered_jobs.sort_values(by='kl_divergence', ascending=True).head(top_n)  # Lower KL divergence means higher similarity
    filtered_jobs['similarity_percentage'] = (100.0 - filtered_jobs['kl_divergence'] * 100.0).round(2)
    filtered_jobs['similarity_percentage'] = filtered_jobs['similarity_percentage'].apply(lambda x: f"{x:.2f}")

    return filtered_jobs[['title_translated', 'employer', 'location', 'url', 'similarity_percentage']]



@app.after_request
def add_header(response):
    response.cache_control.no_store = True
    return response

@app.route('/')
def landing():
    return render_template('landing.html')

@app.route('/search', methods=['GET', 'POST'])
def search():
    if 'user' not in session:
        return redirect(url_for('login'))
    name = users[session['user']]['first_name']
    if request.method == 'POST':
        uploaded_file = request.files['file']
        if uploaded_file.filename != '':
            file_path = f"./{secure_filename(uploaded_file.filename)}"
            uploaded_file.save(file_path)
            extracted_text = extract_text_from_pdf(file_path)
            keyword = request.form.get('keyword', None)
            preferred_provinces = request.form.getlist('preferred_provinces')
            preferred_categories = request.form.getlist('preferred_categories')
            experience_levels = request.form.getlist('experience_levels')
            results_VDAB = find_similar_jobs(vdab_jobs, extracted_text, keyword, preferred_provinces=preferred_provinces, preferred_categories=preferred_categories, experience_levels=experience_levels)
            if results_VDAB.empty:
                return render_template('results_vdab.html', message="No job found for the given criteria.", results_VDAB=pd.DataFrame())
            return render_template('results_vdab.html', results_VDAB=results_VDAB)
    return render_template('search.html', name=name)

@app.route('/about')
def about():
    return render_template('about.html')

@app.route('/contact')
def contact():
    return render_template('contact.html')

@app.route('/terms')
def terms():
    return render_template('terms.html')

@app.route('/privacy')
def privacy():
    return render_template('privacy.html')

@app.route('/cookies')
def cookies():
    return render_template('cookies.html')

@app.route('/testimonies')
def testimonies():
    # Sample testimonies
    testimonies_list = [
        {"name": "Jan De Maeyer", "message": "I found the perfect job thanks to Job Hunter AI! The process was seamless.", "rating": 5},
        {"name": "Bob Smith", "message": "A great tool that helped me land a job quickly. Highly recommended!", "rating": 4},
        {"name": "Katho Van Boven", "message": "Job Hunter AI made my job search so much easier. I love it!", "rating": 5},
        {"name": "Senne De Meester", "message": "I was able to find a job that matched my skills perfectly. Fantastic service!", "rating": 5},
        {"name": "Eve Davis", "message": "A reliable tool that connected me with great job opportunities.", "rating": 4},
        {"name": "Frank Miller", "message": "Job Hunter AI exceeded my expectations. I secured an amazing job in no time!", "rating": 5}
    ]
    return render_template('testimonies.html', testimonies=testimonies_list)

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        email = request.form['email']
        password = request.form['password']
        user = users.get(email)
        if user and user['password'] == password:
            if user.get('confirmed'):
                session['user'] = email
                return redirect(url_for('search'))
            else:
                flash('Email not confirmed. Please check your email for the confirmation link.')
        else:
            flash('Invalid email or password')
    return render_template('login.html')

def send_confirmation_email(email, token):
    subject = "Confirm your email"
    sender = "johhunterugent@gmail.com"
    recipients = [email]
    password = "ttjq dqnr ayyw joon"
    logo_path = os.path.join(app.root_path, 'static', 'job_hunter_logo.PNG')
    confirmation_url = url_for('confirm_email', token=token, _external=True)

    with open(logo_path, 'rb') as f:
        img_data = f.read()

    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    msg['Subject'] = subject

    body = f"""
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
        <div style="max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px;">
            <div style="text-align: center;">
                <img src="cid:logo" alt="Job Hunter AI Logo" style="width: 150px; height: auto;">
            </div>
            <h1 style="color: #ff7f00; text-align: center;">Welcome to Job Hunter AI!</h1>
            <p style="font-size: 16px; line-height: 1.5;">
                Thank you for signing up with Job Hunter AI. To complete your registration, please confirm your email address by clicking the button below:
            </p>
            <div style="text-align: center; margin: 20px 0;">
                <a href="{confirmation_url}" style="background-color: #ff7f00; color: white; padding: 10px 20px; text-decoration: none; border-radius: 5px; font-size: 16px;">
                    Confirm Email
                </a>
            </div>
            <p style="font-size: 16px; line-height: 1.5;">
                If you did not create an account with us, please ignore this email.
            </p>
            <p style="font-size: 16px; line-height: 1.5;">
                Best regards,<br>
                The Job Hunter AI Team
            </p>
        </div>
    </body>
    </html>
    """
    text = MIMEText(body, 'html')
    msg.attach(text)

    image = MIMEImage(img_data, name=os.path.basename(logo_path))
    image.add_header('Content-ID', '<logo>')
    msg.attach(image)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Confirmation email sent!")

@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        first_name = request.form['first_name']
        last_name = request.form['last_name']
        address = request.form['address']
        postal_code = request.form['postal_code']
        email = request.form['email']
        password = request.form['password']
        password_confirmation = request.form['password_confirmation']
        
        if password != password_confirmation:
            flash('Passwords do not match')
        elif email in users:
            flash('Email already registered')
        else:
            token = s.dumps(email, salt='email-confirm')
            users[email] = {
                'first_name': first_name,
                'last_name': last_name,
                'address': address,
                'postal_code': postal_code,
                'password': password,
                'confirmed': False
            }
            with app.app_context():
                send_confirmation_email(email, token)
            flash('A confirmation email has been sent to your email address.')
            return redirect(url_for('login'))
    return render_template('register.html')

@app.route('/confirm/<token>')
def confirm_email(token):
    try:
        email = s.loads(token, salt='email-confirm', max_age=3600)  # 1 hour expiration
        if email in users:
            users[email]['confirmed'] = True
            session['user'] = email
            flash('Email confirmed. You are now logged in.')
            return redirect(url_for('search'))
    except SignatureExpired:
        flash('The confirmation link has expired.')
    return redirect(url_for('register'))

@app.route('/profile', methods=['GET', 'POST'])
def profile():
    if 'user' not in session:
        return redirect(url_for('login'))
    email = session['user']
    user = users[email]
    
    if request.method == 'POST':
        user['first_name'] = request.form['first_name']
        user['last_name'] = request.form['last_name']
        user['address'] = request.form['address']
        user['postal_code'] = request.form['postal_code']
        user['email'] = request.form['email']
        
        # Update session email if changed
        session['user'] = user['email']
        users[user['email']] = users.pop(email)
        flash('Profile updated successfully')
        return redirect(url_for('profile'))
    
    return render_template('profile.html', user=user)

@app.route('/logout')
def logout():
    session.pop('user', None)
    return redirect(url_for('landing'))

if __name__ == '__main__':
    app.run(debug=False)
```

```python
import os
import re
import unicodedata
import pandas as pd
from flask import Flask, render_template, request, redirect, url_for, session, flash
from werkzeug.utils import secure_filename
from bs4 import BeautifulSoup
from itsdangerous import URLSafeTimedSerializer, SignatureExpired
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
import smtplib
import fitz  # PyMuPDF
from sentence_transformers import SentenceTransformer, util
import time
import spacy
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
import torch
import nltk
from nltk.corpus import stopwords
from sklearn.preprocessing import normalize
from scipy.special import rel_entr

app = Flask(__name__, static_url_path='/static', static_folder='static')
app.secret_key = os.urandom(24)  # Secret key for session management

# Email confirmation setup
s = URLSafeTimedSerializer(app.secret_key)

# Temporary in-memory storage for users
users = {
    'admin@admin.com': {
        'first_name': 'Admin',
        'last_name': 'User',
        'address': 'Admin Address',
        'postal_code': '12345',
        'password': 'password',
        'confirmed': True
    }
}
confirmations = {}

# Load VDAB data
vdab_file_path = 'VDAB_categories_provinces_level.csv'
vdab_jobs = pd.read_csv(vdab_file_path)
vdab_jobs['text'] = vdab_jobs['title_translated'].fillna('') + ' ' + vdab_jobs['employer'].fillna('') + ' ' + vdab_jobs['location'].fillna('') + ' ' + vdab_jobs['description_translated'].fillna('')

# Load SBERT model with GPU support if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer('sentence-transformers/all-MiniLM-L12-v2', device=device)

# Function to clean text
def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    # Normalize Unicode characters
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Remove @mentions, URLs, hashtags, non-ASCII characters, and special characters
    text = re.sub(r"@\w+|https?://\S+|#\w+|[^\x00-\x7F]+|[^\w\s]", "", text)
    
    # Remove newline characters, leading/trailing spaces, multiple spaces, and numbers
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"^\s+|\s+$", "", text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"\b\d+\b", "", text)
    
    # Convert to lowercase
    text = text.lower()
    
    return text

# Clean the job descriptions
vdab_jobs['clean_text'] = vdab_jobs['text'].apply(clean_text)

def extract_text_from_pdf(pdf_file_path):
    doc = fitz.open(pdf_file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    text = re.sub(r'\s+', ' ', text)  # Replace multiple whitespace with a single space
    text = text.strip()
    return clean_text(text)

def calculate_kl_divergence(jobs, query_text):
    # Encode jobs in larger batches
    batch_size = 128
    job_embeddings = model.encode(jobs, convert_to_tensor=True, batch_size=batch_size, show_progress_bar=True)
    query_embedding = model.encode([query_text], convert_to_tensor=True, batch_size=1, show_progress_bar=True)

    # Ensure embeddings are valid probability distributions
    job_embeddings = torch.nn.functional.softmax(job_embeddings, dim=1)
    query_embedding = torch.nn.functional.softmax(query_embedding, dim=1)

    # Calculate KL divergence
    kl_divergences = torch.nn.functional.kl_div(query_embedding.log(), job_embeddings, reduction='none').sum(dim=1)
    return kl_divergences

def find_similar_jobs(jobs, query_text, keyword=None, top_n=10, preferred_provinces=None, preferred_categories=None, experience_levels=None):
    if preferred_provinces:
        preferred_provinces = [province.lower() for province in preferred_provinces]

    # Filter VDAB jobs based on keyword, province, categories, and experience level
    filtered_jobs = jobs.copy()
    if keyword:
        filtered_jobs = filtered_jobs[(filtered_jobs['title_translated'].str.contains(keyword, case=False, na=False)) | 
                                      (filtered_jobs['description_translated'].str.contains(keyword, case=False, na=False))]
    if preferred_provinces and 'province' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['province'].str.lower().str.contains('|'.join(preferred_provinces), case=False, na=False)]
    if preferred_categories and 'categories' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['categories'].apply(lambda x: any(cat in preferred_categories for cat in x.split(',')))]
    if experience_levels:
        filtered_jobs = filtered_jobs[filtered_jobs['job_level'].str.contains('|'.join(experience_levels), case=False, na=False)]

    filtered_jobs['clean_text'] = filtered_jobs['clean_text'].apply(clean_text)  # Ensure clean_text is applied

    # Calculate KL divergence for all job embeddings
    kl_divergences = calculate_kl_divergence(filtered_jobs['clean_text'].tolist(), query_text)
    filtered_jobs['kl_divergence'] = kl_divergences.cpu().numpy()
    filtered_jobs = filtered_jobs.sort_values(by='kl_divergence', ascending=True).head(top_n)  # Lower KL divergence means higher similarity
    filtered_jobs['similarity_percentage'] = (100.0 - filtered_jobs['kl_divergence'] * 100.0).round(2)
    filtered_jobs['similarity_percentage'] = filtered_jobs['similarity_percentage'].apply(lambda x: f"{x:.2f}")

    return filtered_jobs[['title_translated', 'employer', 'location', 'url', 'similarity_percentage']]



@app.after_request
def add_header(response):
    response.cache_control.no_store = True
    return response

@app.route('/')
def landing():
    return render_template('landing.html')

@app.route('/search', methods=['GET', 'POST'])
def search():
    if 'user' not in session:
        return redirect(url_for('login'))
    name = users[session['user']]['first_name']
    if request.method == 'POST':
        uploaded_file = request.files['file']
        if uploaded_file.filename != '':
            file_path = f"./{secure_filename(uploaded_file.filename)}"
            uploaded_file.save(file_path)
            
            # Extract text from the uploaded file (assuming it is a PDF)
            extracted_text = extract_text_from_pdf(file_path)
            
            keyword = request.form.get('keyword', None)
            preferred_provinces = request.form.getlist('preferred_provinces')
            preferred_categories = request.form.getlist('preferred_categories')
            experience_levels = request.form.getlist('experience_levels')
            
            # Find similar jobs based on the extracted text and user preferences
            results_VDAB = find_similar_jobs(
                vdab_jobs, 
                extracted_text, 
                keyword=keyword, 
                preferred_provinces=preferred_provinces, 
                preferred_categories=preferred_categories, 
                experience_levels=experience_levels
            )
            
            if results_VDAB.empty:
                flash('No job found for the given criteria.', 'danger')
                return render_template('search.html', name=name)
            
            # Store results in session to be used later for sending email
            session['results_VDAB'] = results_VDAB.to_dict(orient='records')
            return render_template('results_vdab.html', results_VDAB=results_VDAB)
    return render_template('search.html', name=name)


@app.route('/about')
def about():
    return render_template('about.html')

@app.route('/contact')
def contact():
    return render_template('contact.html')

@app.route('/terms')
def terms():
    return render_template('terms.html')

@app.route('/privacy')
def privacy():
    return render_template('privacy.html')

@app.route('/cookies')
def cookies():
    return render_template('cookies.html')

@app.route('/testimonies')
def testimonies():
    # Sample testimonies
    testimonies_list = [
        {"name": "Jan De Maeyer", "message": "I found the perfect job thanks to Job Hunter AI! The process was seamless.", "rating": 5},
        {"name": "Bob Smith", "message": "A great tool that helped me land a job quickly. Highly recommended!", "rating": 4},
        {"name": "Katho Van Boven", "message": "Job Hunter AI made my job search so much easier. I love it!", "rating": 5},
        {"name": "Senne De Meester", "message": "I was able to find a job that matched my skills perfectly. Fantastic service!", "rating": 5},
        {"name": "Eve Davis", "message": "A reliable tool that connected me with great job opportunities.", "rating": 4},
        {"name": "Frank Miller", "message": "Job Hunter AI exceeded my expectations. I secured an amazing job in no time!", "rating": 5}
    ]
    return render_template('testimonies.html', testimonies=testimonies_list)

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        email = request.form['email']
        password = request.form['password']
        user = users.get(email)
        if user and user['password'] == password:
            if user.get('confirmed'):
                session['user'] = email
                return redirect(url_for('search'))
            else:
                flash('Email not confirmed. Please check your email for the confirmation link.')
        else:
            flash('Invalid email or password')
    return render_template('login.html')

def send_confirmation_email(email, token):
    subject = "Confirm your email"
    sender = "johhunterugent@gmail.com"
    recipients = [email]
    password = "ttjq dqnr ayyw joon"
    logo_path = os.path.join(app.root_path, 'static', 'job_hunter_logo.PNG')
    confirmation_url = url_for('confirm_email', token=token, _external=True)

    with open(logo_path, 'rb') as f:
        img_data = f.read()

    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    msg['Subject'] = subject

    body = f"""
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
        <div style="max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px;">
            <div style="text-align: center;">
                <img src="cid:logo" alt="Job Hunter AI Logo" style="width: 150px; height: auto;">
            </div>
            <h1 style="color: #ff7f00; text-align: center;">Welcome to Job Hunter AI!</h1>
            <p style="font-size: 16px; line-height: 1.5;">
                Thank you for signing up with Job Hunter AI. To complete your registration, please confirm your email address by clicking the button below:
            </p>
            <div style="text-align: center; margin: 20px 0;">
                <a href="{confirmation_url}" style="background-color: #ff7f00; color: white; padding: 10px 20px; text-decoration: none; border-radius: 5px; font-size: 16px;">
                    Confirm Email
                </a>
            </div>
            <p style="font-size: 16px; line-height: 1.5;">
                If you did not create an account with us, please ignore this email.
            </p>
            <p style="font-size: 16px; line-height: 1.5;">
                Best regards,<br>
                The Job Hunter AI Team
            </p>
        </div>
    </body>
    </html>
    """
    text = MIMEText(body, 'html')
    msg.attach(text)

    image = MIMEImage(img_data, name=os.path.basename(logo_path))
    image.add_header('Content-ID', '<logo>')
    msg.attach(image)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Confirmation email sent!")

@app.route('/send_email', methods=['POST'])
def send_email():
    if 'user' not in session:
        return redirect(url_for('login'))
    email = session['user']
    user = users[email]
    
    if request.method == 'POST':
        results_VDAB = session.get('results_VDAB')
        if results_VDAB is not None:
            results_VDAB = pd.DataFrame(results_VDAB)
            send_job_results_email(email, results_VDAB)
            flash('Top 5 job results sent to your email.', 'success')
        else:
            flash('No job results to send.', 'danger')
    return jsonify({'redirect': url_for('results', page=1)})

def send_job_results_email(email, results):
    subject = "Your Top Job Matches"
    sender = "johhunterugent@gmail.com"
    recipients = [email]
    password = "ttjq dqnr ayyw joon"

    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    msg['Subject'] = subject

    body = """
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
        <div style="max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px;">
            <h2 style="color: #ff7f00;">Your Top Job Matches</h2>
            <ul style="list-style-type: none; padding: 0;">
    """
    for index, row in results.head(10).iterrows():
        title_translated = row['title_translated'].title()
        employer = row['employer'].title()
        location = row['location'].title()
        similarity_percentage = row['similarity_percentage']
        url = row['url']

        body += f"""
            <li style="margin-bottom: 20px; padding: 10px; border-bottom: 1px solid #ddd;">
                <strong style="font-size: 18px;">{index + 1}. {title_translated}</strong> at {employer} in {location}<br>
                <span style="font-size: 16px;">Similarity: {similarity_percentage}%</span><br>
                <a href="{url}" style="color: #ff7f00;">Apply here</a>
            </li>
        """
    body += """
            </ul>
            <p style="font-size: 16px;">Best regards,<br>The Job Hunter AI Team</p>
        </div>
    </body>
    </html>
    """
    
    text = MIMEText(body, 'html')
    msg.attach(text)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Job results email sent!")
    
    
@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        first_name = request.form['first_name']
        last_name = request.form['last_name']
        address = request.form['address']
        postal_code = request.form['postal_code']
        email = request.form['email']
        password = request.form['password']
        password_confirmation = request.form['password_confirmation']
        
        if password != password_confirmation:
            flash('Passwords do not match')
        elif email in users:
            flash('Email already registered')
        else:
            token = s.dumps(email, salt='email-confirm')
            users[email] = {
                'first_name': first_name,
                'last_name': last_name,
                'address': address,
                'postal_code': postal_code,
                'password': password,
                'confirmed': False
            }
            with app.app_context():
                send_confirmation_email(email, token)
            flash('A confirmation email has been sent to your email address.')
            return redirect(url_for('login'))
    return render_template('register.html')

@app.route('/confirm/<token>')
def confirm_email(token):
    try:
        email = s.loads(token, salt='email-confirm', max_age=3600)  # 1 hour expiration
        if email in users:
            users[email]['confirmed'] = True
            session['user'] = email
            flash('Email confirmed. You are now logged in.')
            return redirect(url_for('search'))
    except SignatureExpired:
        flash('The confirmation link has expired.')
    return redirect(url_for('register'))

@app.route('/profile', methods=['GET', 'POST'])
def profile():
    if 'user' not in session:
        return redirect(url_for('login'))
    email = session['user']
    user = users[email]
    
    if request.method == 'POST':
        user['first_name'] = request.form['first_name']
        user['last_name'] = request.form['last_name']
        user['address'] = request.form['address']
        user['postal_code'] = request.form['postal_code']
        user['email'] = request.form['email']
        
        # Update session email if changed
        session['user'] = user['email']
        users[user['email']] = users.pop(email)
        flash('Profile updated successfully')
        return redirect(url_for('profile'))
    
    return render_template('profile.html', user=user)

@app.route('/logout')
def logout():
    session.pop('user', None)
    return redirect(url_for('landing'))

if __name__ == '__main__':
    app.run(debug=False)
```

```python
import os
import re
import unicodedata
import pandas as pd
from flask import Flask, render_template, request, redirect, url_for, session, flash
from werkzeug.utils import secure_filename
from bs4 import BeautifulSoup
from itsdangerous import URLSafeTimedSerializer, SignatureExpired
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
import smtplib
import fitz  # PyMuPDF
from sentence_transformers import SentenceTransformer, util
import time
import spacy
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertModel, BertTokenizer
import torch
import nltk
from nltk.corpus import stopwords
from sklearn.preprocessing import normalize
from scipy.special import rel_entr

app = Flask(__name__, static_url_path='/static', static_folder='static')
app.secret_key = os.urandom(24)  # Secret key for session management

# Email confirmation setup
s = URLSafeTimedSerializer(app.secret_key)

# Temporary in-memory storage for users
users = {
    'admin@admin.com': {
        'first_name': 'Admin',
        'last_name': 'User',
        'address': 'Admin Address',
        'postal_code': '12345',
        'password': 'Password1$',
        'confirmed': True
    }
}
confirmations = {}

# Load VDAB data
vdab_file_path = 'VDAB_categories_provinces_level.csv'
vdab_jobs = pd.read_csv(vdab_file_path)
vdab_jobs['text'] = vdab_jobs['title_translated'].fillna('') + ' ' + vdab_jobs['employer'].fillna('') + ' ' + vdab_jobs['location'].fillna('') + ' ' + vdab_jobs['description_translated'].fillna('')

# Load SBERT model with GPU support if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer('sentence-transformers/all-MiniLM-L12-v2', device=device)

# Function to clean text
def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    # Normalize Unicode characters
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Remove @mentions, URLs, hashtags, non-ASCII characters, and special characters
    text = re.sub(r"@\w+|https?://\S+|#\w+|[^\x00-\x7F]+|[^\w\s]", "", text)
    
    # Remove newline characters, leading/trailing spaces, multiple spaces, and numbers
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"^\s+|\s+$", "", text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"\b\d+\b", "", text)
    
    # Convert to lowercase
    text = text.lower()
    
    return text

# Clean the job descriptions
vdab_jobs['clean_text'] = vdab_jobs['text'].apply(clean_text)

def extract_text_from_pdf(pdf_file_path):
    doc = fitz.open(pdf_file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    text = re.sub(r'\s+', ' ', text)  # Replace multiple whitespace with a single space
    text = text.strip()
    return clean_text(text)

def calculate_kl_divergence(jobs, query_text):
    # Encode jobs in larger batches
    batch_size = 32
    job_embeddings = model.encode(jobs, convert_to_tensor=True, batch_size=batch_size, show_progress_bar=True)
    query_embedding = model.encode([query_text], convert_to_tensor=True, batch_size=1, show_progress_bar=True)
 
    # Ensure embeddings are valid probability distributions
    job_embeddings = torch.nn.functional.softmax(job_embeddings, dim=1)
    query_embedding = torch.nn.functional.softmax(query_embedding, dim=1)
 
    # Calculate KL divergence
    kl_divergences = torch.nn.functional.kl_div(query_embedding.log(), job_embeddings, reduction='none').sum(dim=1)
    return kl_divergences
 
def normalize_scores(scores):
    min_score = scores.min()
    max_score = scores.max()
    normalized_scores = (scores - min_score) / (max_score - min_score)
    return normalized_scores

def find_similar_jobs(jobs, query_text, keyword=None, top_n=15, preferred_provinces=None, preferred_categories=None, experience_levels=None):
    if preferred_provinces:
        preferred_provinces = [province.lower() for province in preferred_provinces]
 
    # Filter VDAB jobs based on keyword, province, categories, and experience level
    filtered_jobs = jobs.copy()
    if keyword:
        filtered_jobs = filtered_jobs[(filtered_jobs['title_translated'].str.contains(keyword, case=False, na=False)) | 
                                      (filtered_jobs['description_translated'].str.contains(keyword, case=False, na=False))]
    if preferred_provinces and 'province' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['province'].str.lower().str.contains('|'.join(preferred_provinces), case=False, na=False)]
    if preferred_categories and 'categories' in jobs.columns:
        filtered_jobs = filtered_jobs[filtered_jobs['categories'].apply(lambda x: any(cat in preferred_categories for cat in x.split(',')))]
    if experience_levels:
        filtered_jobs = filtered_jobs[filtered_jobs['job_level'].str.contains('|'.join(experience_levels), case=False, na=False)]
 
    filtered_jobs['clean_text'] = filtered_jobs['clean_text'].apply(clean_text)  # Ensure clean_text is applied
 
    # Calculate KL divergence for all job embeddings
    kl_divergences = calculate_kl_divergence(filtered_jobs['clean_text'].tolist(), query_text)
    normalized_kl_divergences = normalize_scores(kl_divergences)
    filtered_jobs['kl_divergence'] = normalized_kl_divergences.cpu().numpy()
    filtered_jobs = filtered_jobs.sort_values(by='kl_divergence', ascending=True).head(top_n)  # Lower KL divergence means higher similarity
    filtered_jobs['similarity_percentage'] = ((1 - filtered_jobs['kl_divergence']) * 100.0).round(2)
    filtered_jobs['similarity_percentage'] = filtered_jobs['similarity_percentage'].apply(lambda x: f"{x:.2f}")
 
    return filtered_jobs[['title_translated', 'employer', 'location', 'url', 'similarity_percentage']]



@app.after_request
def add_header(response):
    response.cache_control.no_store = True
    return response

@app.route('/')
def landing():
    return render_template('landing.html')

@app.route('/search', methods=['GET', 'POST'])
def search():
    if 'user' not in session:
        return redirect(url_for('login'))
    name = users[session['user']]['first_name']
    if request.method == 'POST':
        uploaded_file = request.files['file']
        if uploaded_file.filename != '':
            file_path = f"./{secure_filename(uploaded_file.filename)}"
            uploaded_file.save(file_path)
            
            # Extract text from the uploaded file (assuming it is a PDF)
            extracted_text = extract_text_from_pdf(file_path)
            
            keyword = request.form.get('keyword', None)
            preferred_provinces = request.form.getlist('preferred_provinces')
            preferred_categories = request.form.getlist('preferred_categories')
            experience_levels = request.form.getlist('experience_levels')
            
            # Find similar jobs based on the extracted text and user preferences
            results_VDAB = find_similar_jobs(
                vdab_jobs, 
                extracted_text, 
                keyword=keyword, 
                preferred_provinces=preferred_provinces, 
                preferred_categories=preferred_categories, 
                experience_levels=experience_levels
            )
            
            if results_VDAB.empty:
                flash('No job found for the given criteria.', 'danger')
                return render_template('search.html', name=name)
            
            # Store results in session to be used later for sending email
            session['results_VDAB'] = results_VDAB.to_dict(orient='records')
            return render_template('results_vdab.html', results_VDAB=results_VDAB)
    return render_template('search.html', name=name)


@app.route('/about')
def about():
    return render_template('about.html')

@app.route('/contact')
def contact():
    return render_template('contact.html')

@app.route('/terms')
def terms():
    return render_template('terms.html')

@app.route('/privacy')
def privacy():
    return render_template('privacy.html')

@app.route('/cookies')
def cookies():
    return render_template('cookies.html')

@app.route('/testimonies')
def testimonies():
    # Sample testimonies
    testimonies_list = [
        {"name": "Jan De Maeyer", "message": "I found the perfect job thanks to Job Hunter AI! The process was seamless.", "rating": 5},
        {"name": "Bob Smith", "message": "A great tool that helped me land a job quickly. Highly recommended!", "rating": 4},
        {"name": "Katho Van Boven", "message": "Job Hunter AI made my job search so much easier. I love it!", "rating": 5},
        {"name": "Senne De Meester", "message": "I was able to find a job that matched my skills perfectly. Fantastic service!", "rating": 5},
        {"name": "Eve Davis", "message": "A reliable tool that connected me with great job opportunities.", "rating": 4},
        {"name": "Frank Miller", "message": "Job Hunter AI exceeded my expectations. I secured an amazing job in no time!", "rating": 5}
    ]
    return render_template('testimonies.html', testimonies=testimonies_list)

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        email = request.form['email']
        password = request.form['password']
        user = users.get(email)
        if user and user['password'] == password:
            if user.get('confirmed'):
                session['user'] = email
                return redirect(url_for('search'))
            else:
                flash('Email not confirmed. Please check your email for the confirmation link.')
        else:
            flash('Invalid email or password')
    return render_template('login.html')

def send_confirmation_email(email, token):
    subject = "Confirm your email"
    sender = "johhunterugent@gmail.com"
    recipients = [email]
    password = "ttjq dqnr ayyw joon"
    logo_path = os.path.join(app.root_path, 'static', 'job_hunter_logo.PNG')
    confirmation_url = url_for('confirm_email', token=token, _external=True)

    with open(logo_path, 'rb') as f:
        img_data = f.read()

    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    msg['Subject'] = subject

    body = f"""
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
        <div style="max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px;">
            <div style="text-align: center;">
                <img src="cid:logo" alt="Job Hunter AI Logo" style="width: 150px; height: auto;">
            </div>
            <h1 style="color: #ff7f00; text-align: center;">Welcome to Job Hunter AI!</h1>
            <p style="font-size: 16px; line-height: 1.5;">
                Thank you for signing up with Job Hunter AI. To complete your registration, please confirm your email address by clicking the button below:
            </p>
            <div style="text-align: center; margin: 20px 0;">
                <a href="{confirmation_url}" style="background-color: #ff7f00; color: white; padding: 10px 20px; text-decoration: none; border-radius: 5px; font-size: 16px;">
                    Confirm Email
                </a>
            </div>
            <p style="font-size: 16px; line-height: 1.5;">
                If you did not create an account with us, please ignore this email.
            </p>
            <p style="font-size: 16px; line-height: 1.5;">
                Best regards,<br>
                The Job Hunter AI Team
            </p>
        </div>
    </body>
    </html>
    """
    text = MIMEText(body, 'html')
    msg.attach(text)

    image = MIMEImage(img_data, name=os.path.basename(logo_path))
    image.add_header('Content-ID', '<logo>')
    msg.attach(image)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Confirmation email sent!")

@app.route('/send_email', methods=['POST'])
def send_email():
    if 'user' not in session:
        return redirect(url_for('login'))
    email = session['user']
    user = users[email]
    
    if request.method == 'POST':
        results_VDAB = session.get('results_VDAB')
        if results_VDAB is not None:
            results_VDAB = pd.DataFrame(results_VDAB)
            send_job_results_email(email, results_VDAB)
            flash('Top 5 job results sent to your email.', 'success')
        else:
            flash('No job results to send.', 'danger')
    return jsonify({'redirect': url_for('results', page=1)})

def send_job_results_email(email, results):
    subject = "Your Top Job Matches"
    sender = "johhunterugent@gmail.com"
    recipients = [email]
    password = "ttjq dqnr ayyw joon"

    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    msg['Subject'] = subject

    body = """
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
        <div style="max-width: 600px; margin: auto; padding: 20px; border: 1px solid #ddd; border-radius: 10px;">
            <h2 style="color: #ff7f00;">Your Top Job Matches</h2>
            <ul style="list-style-type: none; padding: 0;">
    """
    for index, row in results.head(10).iterrows():
        title_translated = row['title_translated'].title()
        employer = row['employer'].title()
        location = row['location'].title()
        similarity_percentage = row['similarity_percentage']
        url = row['url']

        body += f"""
            <li style="margin-bottom: 20px; padding: 10px; border-bottom: 1px solid #ddd;">
                <strong style="font-size: 18px;">{index + 1}. {title_translated}</strong> at {employer} in {location}<br>
                <span style="font-size: 16px;">Similarity: {similarity_percentage}%</span><br>
                <a href="{url}" style="color: #ff7f00;">Apply here</a>
            </li>
        """
    body += """
            </ul>
            <p style="font-size: 16px;">Best regards,<br>The Job Hunter AI Team</p>
        </div>
    </body>
    </html>
    """
    
    text = MIMEText(body, 'html')
    msg.attach(text)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Job results email sent!")
    
    
@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        first_name = request.form['first_name']
        last_name = request.form['last_name']
        address = request.form['address']
        postal_code = request.form['postal_code']
        email = request.form['email']
        password = request.form['password']
        password_confirmation = request.form['password_confirmation']
        
        # Check if passwords match
        if password != password_confirmation:
            flash('Passwords do not match')
        # Check if email is already registered
        elif email in users:
            flash('Email already registered')
        # Validate password with regex
        elif not re.match(r'^(?=.*[A-Za-z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!%*#?&]{8,}$', password):
            flash('Password must be at least 8 characters long and include at least one number and one special character.')
        else:
            token = s.dumps(email, salt='email-confirm')
            users[email] = {
                'first_name': first_name,
                'last_name': last_name,
                'address': address,
                'postal_code': postal_code,
                'password': password,
                'confirmed': False
            }
            with app.app_context():
                send_confirmation_email(email, token)
            flash('A confirmation email has been sent to your email address.')
            return redirect(url_for('login'))
    return render_template('register.html')


@app.route('/confirm/<token>')
def confirm_email(token):
    try:
        email = s.loads(token, salt='email-confirm', max_age=3600)  # 1 hour expiration
        if email in users:
            users[email]['confirmed'] = True
            session['user'] = email
            flash('Email confirmed. You are now logged in.')
            return redirect(url_for('search'))
    except SignatureExpired:
        flash('The confirmation link has expired.')
    return redirect(url_for('register'))

@app.route('/profile', methods=['GET', 'POST'])
def profile():
    if 'user' not in session:
        return redirect(url_for('login'))
    email = session['user']
    user = users[email]
    
    if request.method == 'POST':
        user['first_name'] = request.form['first_name']
        user['last_name'] = request.form['last_name']
        user['address'] = request.form['address']
        user['postal_code'] = request.form['postal_code']
        user['email'] = request.form['email']
        
        # Update session email if changed
        session['user'] = user['email']
        users[user['email']] = users.pop(email)
        flash('Profile updated successfully')
        return redirect(url_for('profile'))
    
    return render_template('profile.html', user=user)

@app.route('/logout')
def logout():
    session.pop('user', None)
    return redirect(url_for('landing'))

if __name__ == '__main__':
    app.run(debug=False)
```
