```python
import pandas as pd
```

```python
# Load VDAB data
vdab_file_path = 'C:/Users/Cool_/Documents/GitHub/PPA/hr_tool_irakli/VDAB_data_translated_provinces.csv'
vdab_jobs = pd.read_csv(vdab_file_path)
```

```python
# Define the categories and corresponding general keywords
categories = {
    "Administration": ["administration", "admin", "secretary", "clerk", "office", "assistant"],
    "Agriculture & fishing": ["farmer", "fisherman", "agriculture", "fishing"],
    "Construction & real estate": ["construction", "real estate", "builder", "architect", "civil engineer", "property", "tiler", "ground worker"],
    "Creative jobs": ["creative", "designer", "artist", "graphic", "writer", "illustrator", "marketer"],
    "Education & training": ["teacher", "instructor", "trainer", "professor", "educator", "education"],
    "Executive board & management": ["manager", "executive", "director", "ceo", "cfo", "coo", "leader", "supervisor", "coordinator"],
    "Finance, Insurance & law": ["finance", "financial", "analyst", "lawyer", "insurance", "accountant", "auditor", "controller", "tax"],
    "Government": ["government", "public", "municipal", "council"],
    "Health & body care": ["health", "care", "nurse", "doctor", "therapist", "dentist", "pharmacist", "medical", "beautician"],
    "Hotel, catering, tourism, maintenance": ["hotel", "catering", "tourism", "chef", "cook", "cleaning", "maintenance", "housekeeping", "waiter", "hospitality", "baker", "pastry"],
    "Human resources": ["HR", "human resources", "recruiter", "talent", "recruitment", "payroll"],
    "ICT": ["IT", "information technology", "developer", "engineer", "programmer", "technician", "support", "consultant", "analyst"],
    "Logistics, transport & purchase": ["logistics", "transport", "purchaser", "driver", "warehouse", "order picker", "dispatcher", "stock"],
    "Marketing & communication": ["marketing", "communication", "marketer", "digital"],
    "Research & development": ["research", "development", "R&D"],
    "Sales": ["sales", "salesman", "salesperson", "saleswoman", "account manager"],
    "Social, cultural & sports": ["social", "cultural", "sports", "coach", "assistant"],
    "Surveillance & security": ["security", "surveillance", "guard"],
    "Technics, engineering & production": ["technician", "engineer", "production", "operator", "mechanic", "electrician", "welder", "assembler", "supervisor"],
    "Trades": ["trade", "trader", "shop", "store", "cashier", "installer", "mechanic", "repair"]
}

# Function to categorize jobs based on description
def categorize_job(description):
    for category, keywords in categories.items():
        if any(keyword in description.lower() for keyword in keywords):
            return category
    return "Uncategorized"

# Apply the categorization
vdab_jobs['category'] = vdab_jobs['description_translated'].apply(categorize_job)

# Display the value counts for the categories
category_counts = vdab_jobs['category'].value_counts()
```
